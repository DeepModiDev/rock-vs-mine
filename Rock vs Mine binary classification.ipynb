### Problem statement (Beginner Level)

In this problem we have the sonar data collected from the submarine. This data contains the values related to Rock and mine. Basically, what we have to do is we need to find whether the object under the water is rock or mine based on the sonar value. This is a binary classification problem.

We can solve this problem using Logistic Regression and Neural Networks.

### Seperates the features and target class columns

```python
X = submarine_data.drop(columns=60, axis=1)
y = submarine_data[60]
```

```python
print("Features: ")
X.head()
```

```python
print("Target labels: ")
y.head()
```

### Now split dataset into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Model Training

### LogisticRegression model training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

### LogisticRegression model evaluation

```python
x_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on test data: ",test_accuracy)
```

### Neural Network model training

### First apply data transformation on targeted class 'y'

```python
labels = ['Rock','Mines']
```

```python
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
```

```python
encoded_Y
```

```python
new_train_X, new_test_X, new_train_y, new_test_y = train_test_split(X,encoded_Y, test_size=0.1, stratify=encoded_Y, random_state=1)
```

```python
nn_model = Sequential([
    tf.keras.layers.Input(shape=(60,)),
    tf.keras.layers.Dense(units=64, activation="relu"),
    tf.keras.layers.Dense(units=1, activation="sigmoid"),
])
```

```python
nn_model.summary()
```

```python
nn_model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
history = nn_model.fit(new_train_X,new_train_y,epochs=500, verbose=1)
```

```python
nn_testing_results = nn_model.predict(new_test_X)
```

```python
rounded = [int(round(x[0])) for x in nn_testing_results]
print(rounded)
```

```python
nn_test_accuracy = accuracy_score(rounded,new_test_y)
print("NN Model's accuracy on test data: ",nn_test_accuracy)
```
